### Зачем тестировать гипотезы?
Мы живем в мире неопределенности. Нам приходится принимать решения, влияющие на наше счастье и благосостояние, в условиях неполной информации.

Например, хипстер Ваня начинает бизнес, инвестируя в него свои время и деньги, не зная, сколько он сможет на этом заработать; отправляясь в спортивное путешествие по Европе на велосипеде, он не знает, повезет ли ему с погодой; а делая предложение Маше, он не знает, согласится она или нет.

Поскольку информация заведомо неполная, мы не можем избежать ошибок, но нам нужен инструмент, который бы позволил нам свести к минимуму количество случаев, когда мы ошибаемся.

Тестирование гипотез — это именно такой инструмент, инструмент принятия оптимальных решений. Интуитивно мы используем его в жизни постоянно.

### Пример
Маша влюбилась в нашего хипстера Ваню. Она не знает, взаимно ли это, поэтому дожидается 14 февраля и видит, что он не дарит ей валентинку (и не проявляет других знаков внимания). После чего принимает решение, что Ваня ее не любит.

Может показаться, что ничего особенного в этом нет и статистика здесь точно ни при чем. Однако эта ситуация наглядно иллюстрирует, как мы тестируем гипотезы в повседневной жизни. Разберем по шагам:

#### 1

Маша пребывает в неполной информации, она не знает, что верно: гипотеза о том, что Ваня ее любит, или же о том, что не любит.

Она формулирует гипотезы:

$H_0$: «Ваня любит меня» (нулевая гипотеза)

$H_A$: «Ваня не любит меня» (альтернативная гипотеза)

#### 2
Чтобы принять решение, в какое из утверждений верить, она наблюдает за его поведением, другими словами, собирает наблюдения. 

#### 3
Получив наблюдения, Маша делает мысленный эксперимент: допустим, $H_0$ верна. Чего тогда ей следует ожидать от Вани на 14 февраля? Скорее всего, думает Маша, он должен за мной поухаживать: подарить цветы, пригласить на ужин или хотя бы найти способ передать мне валентинку. Исходя из этих предположений, Маша оценивает, насколько ожидаемым является наблюдаемое поведение Вани, то есть вероятность того, что Ваня повёл бы себя именно так, будь гипотеза о любви верна.

Мысленный эксперимент: Если $H_0$ верна, как должен вести себя Ваня?

#### 4
Далее, если эта вероятность слишком мала («Я не могу себе представить, чтобы влюбленный человек даже не поздравил с 14 февраля!»), гипотеза о любви отвергается в пользу альтернативной гипотезы (что не любит). А если поведение таково, что эта вероятность не мала (скажем, Ваня подарил открытку и позвал вечером пройти вместе квест) – гипотеза о любви не отвергается.

Если p-value мала (< уровня значимости $\alpha$), $H_0$ отвергается в пользу $H_A$. Если вероятность не мала — не отвергается.

#### 5
Таким образом, если Ваня не подарил открытку, Маша решает, что он ее не любит.

### Термины
Основная гипотеза, которую мы тестируем, называется нулевой, а противная ей — альтернативной. В английском языке они называются соответственно *null hypothesis* и *alternative hypothesis*, поэтому обозначаются часто $H_0$ и $H_1$.

Вероятность полученных наблюдений, оценённая исходя из того, что нулевая гипотеза верна, называется p-value.

А число, при котором мы считаем, что p-value достаточно мало, чтобы отвергнуть нулевую гипотезу, называется уровнем значимости α. Чаще всего это либо 1%, либо 5%, либо 10%.

## 4.2. «Это не точно!», или ошибки при тестировании

Какое бы решение мы ни приняли в результате тестирования гипотезы, к нему всегда нужно добавить: но это не точно!

Наши решения основаны на неполной информации, то есть лишь на некотором количестве наблюдений, поэтому есть шанс ошибиться!

### Типы ошибок
#### 1.Ошибка I рода
Эта ошибка может возникнуть, когда отвергли нулевую гипотезу, которая была верна.

Верна $H_0$, но ее отвергли

P(ошибка I)=$\alpha$

Вероятность ошибки I рода обозначается буквой $\alpha$.

#### 2. Ошибка II рода
Эта ошибка может возникнуть, когда верна альтернативная гипотеза, но при этом нулевую гипотезу не отвергли.

Верна $H_A$, но $H_0$ не отвергли

P(ошибка II)=$\beta$

Вероятность ошибки II рода обозначается буквой $\beta$.

### Почему вероятность ошибки  рода совпадает с уровнем значимости ?
Давайте подумаем. Допустим, нулевая гипотеза верна. В этом случае есть шанс, что мы получим очень необычную выборку, в которой среднее значение будет серьезно меньше того, что мы ожидаем по нулевой гипотезе. Мы посчитаем p-value, и оно окажется очень маленьким. Поскольку мы, принимая решение, сравниваем его с , то  и будет вероятностью того, что мы ошибемся.

Итак, что бы вы ни решили при тестировании гипотезы, возможно, вы ошиблись. Но скорее всего, это не так!

Вернемся к примеру любви Маши к Ване.

### Любовные ошибки
В этой ситуации **ошибкой I рода** будет принятое Машей решение о том, что Ваня ее не любит. Возможно, Ваня ее любит, но случилось нечто непредвиденное: он запутался в датах, забыл подарок дома, его съела собака или украли инопланетяне.

**Ошибкой II рода** будет принятое Машей решение о том, что Ваня ее любит: например, когда Ваня подарил ей открытку, и она решила, что он ее любит. Это необязательно может быть так: Ваня просто может любить дарить открыточки и не придавать этому поступку особое значение. А для Маши это был критерий для принятия решений.

Далее мы рассмотрим несколько примеров тестирования гипотез на конкретных числах. Все три примера будут касаться тестирования гипотез об истинном среднем значении $\mu$.

### Формулируем гипотезу
Мэр некоторого города в средней полосе России утверждает, что средняя заработная плата в городе составляет 55000 рублей. Мы хотим проверить это утверждение.

Записываем данные
$\mu=55000$ рублей?

X — зарплата жителя города

На случайной выборке из 40 опрошенных человек мы посчитали среднее, оно получилось равным 35000 рублей.

Выборка: $n=40, \bar{X}=35000$

Из опубликованных статистических данных мы узнали, что истинное стандартное отклонение заработных плат составляет 15500 рублей. 
$$\sigma = 15500$$
Первый шаг для тестирования — постановка нулевой и альтернативной гипотез.

Нулевая гипотеза — то, от чего мы отталкиваемся, поэтому там всегда будет равенство.

В данном случае: $H_{0}: \mu = 55000$

Альтернативная гипотеза содержит то, что мы хотим проверить, и это всегда неравенство, которое выражает направление нашего подозрения или недоверия к нулевой гипотезе.

Крошка Мю сейчас поможет нам поставить альтернативную гипотезу.

![alt_img](pic/stat-4-2-1.png)

Кажется, она не доверяет мэру нашего города и подозревает, что истинное среднее $\mu$ меньше, чем 55000.

Значит, именно это мы и запишем в альтернативную гипотезу:
$$H_{A}: \mu < 55000$$

## 4.3. Гипотеза об уровне зарплат

### Мысленный эксперимент
Что ж, мы сформулировали гипотезы и собрали наблюдения. Следующий шаг — мысленный эксперимент.

Допустим, нулевая гипотеза верна. Какое значение среднего дохода на выборке мы ожидаем получить?

Что ж, выборка большая (n > 30) и случайная, значит по центральной предельной теореме выборочное среднее распределено приблизительно нормально вот с такими параметрами:
$$\bar{X} \approx N(55000, \frac{15500}{\sqrt{40}})$$
Если мы предполагаем, что нулевая гипотеза верна, мы подставляем известные значения $\sigma$ и n, и также $\mu = 55000$.

Значит, на этой выборке мы в среднем ожидаем получить выборочное среднее, равное 55000. Насколько неожиданным в таком случае является выборочное среднее в 35000 рублей?

### Считаем z-статистику
Чтобы дать оценку нашему наблюдаемому значению, давайте посчитаем z-статистику.

Напомним, что z-статистика показывает, на сколько стандартных отклонений нормальная случайная величина отклонилась от своего среднего. Чем больше это число по модулю, тем дальше наблюдение от того, что мы ожидали получить, и тем меньше вероятность такого наблюдения.

Итак, модуль z-статистики показывает, насколько неожиданным или неправдоподобным является наблюдаемая выборка, предполагая, что нулевая гипотеза верна.

Считаем z-статистику:
    $$z_{st} = \frac{\bar{x} - \mu_{0} }{\sigma / \sqrt{n}} = \frac{35000-55000}{\frac{15500}{\sqrt{40}}} = - 8.16$$
Ого! Это означает, что если нулевая гипотеза верна, то наблюдаемое выборочное значение отклонилось от ожидаемого среднего () на более чем 8 стандартных отклонений. Что ж, с точки зрения нулевой гипотезы такое наблюдение кажется достаточно неожиданным. Но насколько неожиданным?

Как нам оценить, достаточно ли это далеко от ожидания, чтобы отвергнуть нулевую гипотезу?

### p-value
Это можно оценить при помощи p-value. p-value — это вероятность получить такую неожиданную (или еще более неожиданную) выборку при условии, что нулевая гипотеза верна.

Какую выборку можно считать неожиданной или странной? Что ж, по нулевой гипотезе мы считаем, что $\mu$ равно 55000. В таком случае X среднее должно быть близко к этому значению. Мы же наблюдаем значение 35000, и это уже достаточно странно. Что же считать ещё более странной выборкой?

Альтернативная гипотеза всегда показывает нам направление нашего подозрения по отношению к нулевой гипотезе: мы подозреваем, что μ меньше 55000 рублей. Если это так, то рассчитанная по нашей формуле z-статистика чаще всего будет давать отрицательное значение, поскольку чаще всего X среднее будет меньше, чем $\mu_0$.

Значит, p-value в нашем случае будет вероятностью того, что выборочное X среднее равно или меньше 55000. Это то же самое, что вероятность того, что Z будет равен или меньше -8.16.

Находим p-value: 

p-value $= P(Z \leq -8.16) = 1.7 \cdot  10^{-16}$

О том, как посчитать  p-value для z-статистики при помощи python, вы узнаете из раздела Практика: z-статистика, t-статистика.

Эта вероятность очень мала. Но решение о том, что она слишком мала, мы принимаем путем сравнения p-value и $\alpha$.

Допустим, $\alpha = 0.05$

Сравниваем p-value и $\alpha$:

p-value $=0.00000000000000017$ и $\alpha = 0.05$
 
p-value < $\alpha$

P-value оказалась меньше $\alpha$, что дает нам право отвергнуть нулевую гипотезу в пользу альтернативной: мэр города нам врет и $\mu$ меньше 55000 рублей.

### Графическое решение
Рассмотрим, как это решение выглядит графически.

#### 1
Мы тестируем вот такую пару гипотез: $\mu = 55000$ и $\mu < 55000$

$\mu_0$ — это стандартное обозначение для значения, которое указано в нулевой гипотезе.

#### 2
Так как выборка большая и случайная, выборочное среднее распределено приблизительно стандартно нормально.

#### 3
Мысленный эксперимент. Допустим, нулевая гипотеза верна, тогда:
$$\mu =\mu_{0}$$
$$Z = \frac{\bar{X} - \mu_{0} }{\sigma / \sqrt{n}} \approx N(0, 1)$$
Тогда мы ожидаем, что выражение описывается стандартным нормальным распределением. То есть если $H_0$ верна, и мы возьмем много случайных выборок, посчитаем на каждой X среднее и z-статистику, то результаты будут распределены вот так:
![alt_img](pic/stat-4-3-1.png)

#### 4
Ожидания построили, теперь оценим, насколько неожиданно то, что мы наблюдаем.

Мы получили $z_{st} = -8.16$, то есть на 8 стандартных отклонений ниже того, что мы ожидали по нулевой гипотезе. Как на графике посчитать вероятность этого, то есть p-value?

p-value – это вероятность того, что Z окажется меньше $z_{st} = -8.16$. Ситуация, при которой  будет получаться чаще всего отрицательной, будет происходить в левом хвосте распределения.

Найти эту вероятность мы можем как площадь под кривой.

#### 5
Насколько это мало? Чтобы это оценить, нам надо сравнить эту площадь с уровнем значимости $\alpha$. Допустим, $\alpha=0.05$. Как его изобразить? Очевидно, что это площадь слева, так как именно отрицательные значения z нам кажутся подозрительными.

Кстати, значение $z_cr$ в этой точке равно -1.645. Мы легко можем рассчитать значение, используя калькулятор - https://planetcalc.ru/7803/.

Мы можем прийти к решению, сравнивая площадь p-value с площадью $\alpha$. Так как p-value меньше критически низкого значения $\alpha$, нулевая гипотеза отвергается.

К этому же выводу можно прийти, сравнивая $z_st$ и $z_cr$. Так как $z_st$ оказалась дальше от ожидаемого значения 0, чем нам позволяет $\alpha$, то нулевая гипотеза отвергается.

## 4.4. Гипотеза об уровне загрязнения

### Гипотеза об уровне загрязнения

Где-то на окраине нашего города строится аккумуляторный завод, но применение устаревших технологий предполагает выброс вредных веществ в атмосферу. Наш мэр утверждает, что среднее содержание вредных веществ в атмосфере не превысит 25ppm, то есть будет находиться в пределах допустимой нормы. Мы хотим проверить это утверждение.

### Записываем условия
$μ= 25 ppm$?

X — содержание вредных веществ в атмосфере

Мы делаем 20 независимых замеров и считаем $\bar{X}$, которое получилось равно 37ppm.

Выборка: $n=20, \bar{X} =37$

Допустим, нам также известно, что при таких измерениях стандартное отклонение среднего равно 14, и что X распределен нормально.
$$\sigma = 14 ppm, X \sim N$$

#### 1. Формулируем нулевую и альтернативную гипотезы
Мы подозреваем, что мэр нашего города снова лжет и $\mu$ превышает названное им значение.

$H_{0}:\mu = 25$

$H_{A}:\mu > 25$

#### 2. Собираем наблюдения

Поскольку X распределен нормально, а выборка случайная и большая, X среднее распределено тоже нормально.
$$n = 20, \bar{X} = 37, \sigma = 14ppm, X\sim N$$
$$\bar{X}\approx N\left ( \mu , \frac{\sigma }{\sqrt{n}} \right )$$

#### 3. Мысленный эксперимент

Допустим, нулевая гипотеза верна. В таком случае $μ= 25 ppm$, и тогда X среднее будет распределено с такими параметрами:
$$H_{0}: \bar{X}\approx N(25, \frac{14}{\sqrt{20}})$$

#### 4. Считаем zst и p-value
Мы оцениваем, насколько неожиданны наши предположения при условии, что нулевая гипотеза верна. Для этого рассчитываем z-статистику, подставляя значения в формулу:
$$Z_{st} = \frac{\bar{x} - \mu_{0} }{\sigma /\sqrt{n}} = \frac{37-25}{\frac{14}{\sqrt{20}}} = 3.83$$

Значение 3.83 означает, что наблюдаемое выборочное среднее находится на расстоянии почти 4 стандартных отклонений от ожидаемого по нулевой гипотезе значения.

Теперь нам нужно оценить, насколько это наблюдение неожиданно, с помощью p-value.

Наша альтернативная гипотеза говорит нам, что $\mu$ больше 25. В этой ситуации X среднее будет больше 25, а Z окажется положительным. Поэтому p-value в нашем случае считается так:
$$p-value = P(\bar{X}\geq 37\mid H_{0}) = P(Z\geq 3.83) = 0.000064$$

#### 5. Сравниваем p-value и $\alpha$

$\alpha = 0.05$, p-value = 0.000064

p-value < $\alpha$

Наблюдаемые значения на выборке исходя из предположения, что нулевая гипотеза верна, оказались слишком маленькие. Это означает, что нулевая гипотеза неправдоподобна, и мы ее отвергаем.

Что ж, кажется, мэр нас снова обманул, и после запуска аккумуляторного завода содержание вредных веществ в атмосфере превысит допустимые нормы.

## 4.5. Гипотеза об удое молока

Наш хипстер Ваня решил открыть дело по производству сыра бри. В процессе расчетов ему понадобилось понять, сколько коров ему нужно, чтобы производить необходимое количество молока. Ему требуется информация о среднем удое с коровы.

Где-то он обнаружил данные, что в среднем корова дает 12 л молока, но ему хочется проверить это утверждение. Он собрал информацию у находящихся неподалеку фермеров, взял выборку из 25 коров и посчитал выборочное среднее, которое оказалось равно 10.8 л.

Также он знает, что стандартное отклонение равно 5 л, и что X распределен приблизительно нормально.

### Записываем условия:
$\mu = 12$ л

$\bar{X}$ — суточный удой молока с коровы

$\sigma = 5$ л, $X\sim N$
Выборка: $n = 25, \bar{X} = 10.8$ л

#### 1. Формулируем гипотезу
У нас нет никаких подозрений, поэтому наша альтернативная гипотеза состоит в том, что $\mu$ не равно 12. Такая гипотеза называется двусторонней.
$$H_{0}: \mu = 12$$
$$H_{A}: \mu \neq  12$$

#### 2. Собираем наблюдения
$$n = 25, \bar{X} = 10.8, \sigma = 5, X\sim N$$
$$\bar{X} \approx N(\mu, \frac{\sigma }{\sqrt{n}})$$
 
#### 3. Мысленный эксперимент

По нулевой гипотезе $\mu = 12$ л, и мы можем его подставить в предполагаемое распределение X среднего.
$$\bar{X}\approx N(12,\frac{5}{\sqrt{25}})$$

#### 4. Считаем $Z_st$  и p-value
Насколько далеко лежит наблюдаемое значение от ожидаемого по нулевой гипотезе? Считаем z-статистику:
$$z_{st} = \frac{\bar{x}-\mu_{0} }{\sigma /\sqrt{n}} = \frac{10.8 - 12}{\frac{5}{\sqrt{25}}} = -1.2$$

Считаем p-value:

p-value $= P(\left | Z \right |\geq 1.2) = 2P(Z\leq -1.2)=2*0.115=0.23$

Чтобы понять, -1.2 — слишком или не слишком далеко, используем p-value. Значение p-value покажет нам вероятность того, что выборка окажется ещё более странной при условии, что нулевая гипотеза верна. Но какую выборку считать странной, если наша альтернативная гипотеза двусторонняя? Это значит, что любое отклонение от ожидаемого значения будет казаться нам подозрительным. Поэтому при расчете p-value мы будем считать вероятность по модулю (ведь распределение симметричное).

#### 5. Сравниваем p-value и α.
$\alpha = 0.05$

p-value$= 0.23 > \alpha = 0.05$

Таким образом, наблюдаемые значения не кажутся нам достаточно подозрительными, чтобы отвергнуть нулевую гипотезу. Мы остаемся при убеждении, что $\mu = 12$ л.

## 4.6. Другие тестовые статистики

### Как считать p-value?
Давайте обобщим эти три примера. В них мы обнаружили, что бывают разные типы альтернативных гипотез. Это важно знать, поскольку p-value для этих типов рассчитывается по-разному.
- Если альтернативная гипотеза содержит знак <, то p-value рассчитывается как вероятность того, что Z будет меньше наблюдаемой z-статистики.
- Если альтернативная гипотеза содержит знак >, то p-value рассчитывается как вероятность того, что Z будет больше наблюдаемой z-статистики.
- Если альтернативная гипотеза содержит знак неравенства (двусторонняя альтернатива), то p-value рассчитывается как вероятность того, что Z будет меньше наблюдаемой z-статистики. Другими словами, он будет равен двум площадям p-value для односторонней альтернативы.

### Тестирование гипотез о среднем $\mu$
Мы тестировали гипотезы, зная значение истинного стандартного отклонения $\sigma$. Однако в реальной жизни это весьма экзотический случай: если мы знаем $\sigma$, мы знаем и $\mu$, а если мы знаем $\mu$, то нам не нужно делать тест.

В реальной жизни мы чаще всего мы заменяем значение истинного стандартного отклонения $\sigma$ на значение выборочного стандартного отклонения S.

Мы знаем, что при этой замене распределение сохраняет свои свойства, но разброс возможных значений становится больше, и z-статистика превращается в t-статистику (которая следует распределению Стьюдента).

Поэтому для тестирования гипотез о среднем $\mu$ у нас есть две формулы:

1. Когда $\sigma$ известно, мы используем z-статистику $$z_{st} = \frac{\bar{x} - \mu_{0}}{\sigma /\sqrt{n}}$$
2. Когда $\sigma$ неизвестно, мы используем t-статистику $$t_{st} = \frac{\bar{x} - \mu_{0}}{s /\sqrt{n}}$$
А количество степеней свободы считается следующим образом: $$k=n-1$$
В обоих случаях важно убедиться в двух вещах:

выборка взята случайно
выборка достаточно велика или X распределен нормально $(X\sim N)$

### Тестирование гипотез о пропорции $\hat {p}$
Мы знаем, что если выборка велика и взята случайно, по центральной предельной теореме выборочная пропорция распределена приблизительно нормально. Любую нормальную величину можно превратить в стандартную нормальную величину: достаточно вычесть из нее среднее p и поделить на стандартное отклонение. 
$$\hat{p} \approx N (p, \sqrt{\frac{p(1-p)}{n}}) \rightarrow \frac{\hat{p} - p}{\sqrt{\frac{p(1-p)}{n}}}\approx Z\sim N(0,1)$$

При тестировании гипотез на третьем шаге мы предположим, что нулевая гипотеза верна. Поэтому при подсчете z-статистики вместо значения p мы будем использовать $p_0$ (то значение пропорции, которое будет указано в нулевой гипотезе).
$$z_{st} = \frac{\hat{p} - p_{0}}{\sqrt{\frac{p_{0}(1-p_{0})}{n}}}$$
В обоих случаях важно убедиться в важных вещах:
- выборка взята случайно
- выборка достаточно велика
$$(n\hat{p} \geq 5,\: n(1-\hat{p})\geq 5)$$
