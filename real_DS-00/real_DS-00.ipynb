## 0.1. О дивный, новый мир!

### Аналитик
Задача аналитика — помогать бизнесу принимать решения на основе цифр.

Пример. У руководителя проекта возник вопрос: почему упал retention rate на сайте?
Аналитик проводит исследование и выдвигает гипотезу: это может быть связано с сезонностью, возможно, повлияли изменения в интерфейсе (неудобная корзина), окончание распродаж или программы лояльности. Далее аналитик решает, какие данные нужно собрать, чтобы подтвердить или опровергнуть каждую из гипотез. Затем на основе данных выдаёт ответ бизнесу и пишет рекомендации о том, что нужно сделать.
Алгоритм действий
Получить бизнес задачу ⇒ проанализировать ее ⇒ сформировать гипотезу для Data Science Eengineer ⇒ написать отчёт с выводами и рекомендациями.

Какими навыками должен обладать аналитик?
 - Обязательно уметь работать с базами данных (в том числе знать SQL).
 - Знать один из языков программирования: Python, Scala или R.
 - Иметь опыт работы с SAS (это требование актуально в основном для банковской среды).

### Математик-программист (Data Science Engineer)
Задача математика-программиста — построение математических моделей, на которых будет обучаться компьютер.

Пример. Есть онлайн-магазин, которому нужны рекомендательные системы, какие товары чаще покупаются в сочетании друг с другом. Математик-программист договаривается с бизнесом о метрике, которую они будут измерять (например, конверсию в продажи), после чего собирает данные, пишет код, который формирует обучающие выборки для машинного обучения, строит модели, пытается оптимизировать метрики и параметры самой модели. 
Алгоритм действий
Очистить данные ⇒ написать код, который обрабатывает данные ⇒ построить модель ⇒ получить метрики, которые адекватно соотносятся с бизнес-запросом.

Какими навыками должен обладать математик-программист?
 - Хорошо знать разделы математики и статистики, использующиеся в машинном обучении.
 - Обладать навыками программирования: Python (библиотеки, с помощью которых реализуются алгоритмы машинного обучения), Scala (востребован, потому что сейчас очень популярен Spark), R.
 - Иметь опыт работы со Spark, Hadoop, Kafka — инструментами по работе с большими данными.

### Инженер данных (Data Engineer)
Задача дата инженера — сделать так, чтобы данные собирались, комбинировались, складывались в хранилище, сохранялись и были доступны для работы. К этим задачам еще обычно добавляется внедрение модели в производство.

Пример. Если дата инженер понимает, что он собирает данные, с которыми потом будут работать BI-аналитики, то для их хранения он выбирает какую-то простую реляционную базу данных, принимая во внимание, что все аналитики знают SQL. Однако если есть информация, что проект будет масштабироваться, и в будущем в нём будут использоваться более сложные модели данных, инженер изначально будет ориентироваться на более сложные системы: Postgres, Cassandra и т. д.
Алгоритм действий
Подумать, в каком виде стоит хранить данные (очистить данные), кто с ними потом будет работать и как ⇒ выбрать хранилище ⇒ сделать так, чтобы данные дублировались и не терялись.

Какими навыками должен обладать инженер данных (data engineer)?
 - Иметь навыки программирования на Python, Java или Scala.
 - Знать реляционную модель данных, её преимуществ и ограничений.
 - Знать другие (NoSQL) модели. иметь навыки работы с Hadoop.
 - Знать парадигму map reduce.
 - Знать распределённые файловые системы типа HDFS и BigTable. 
 - Иметь архитектурное мышление и уметь проектировать целые конвейеры обработки данных. 

### Учёные (Data Science Researcher)
Специалисты, работающие в области Data Science Research, придумывают новые алгоритмы машинного обучения, ищут новые подходы к работе с данными, генерируют новые гипотезы, предлагают новые математические модели. А ещё они пишут статьи, выступают на конференциях, одним словом, двигают науку вверх.

### Заполняем резюме

Для создания резюме, которое может заинтересовать рекрутёра или потенциального работодателя, мы рекомендуем использовать шаблон с сайта HeadHunter

Ещё несколько важных трюков:
 - Если, работая в одной компании, вы переходили с одной должности на другую (профессионально развивались), то опыт внутри компании не стоит разбивать на периоды, а лучше описать его единым блоком, с указанием времени начала о окончания работы в компании, и наименования последней должности.
 - Резюме всегда украшают достижения и KPI. Такая информация помогает работодателю чётко представить, что вы можете принести его компании. При этом не стоит путать достижения со своими функциональными обязанностями. 
 - Если у вас есть какие-то ссылки на ваши выступления, статьи, преподавательскую деятельность или тот же профиль на Kaggle или GitHub, эта информация отлично впишется в блок «обо мне».

## 0.5. Космический карьерный трек

Чтобы достигнуть мечты, нужно иметь не только точные конечные координаты, ориентироваться по «светилам» и уверенно перемещаться «в открытом космосе», но и построить чёткий трек достижения цели.

Поэтому первым делом давайте сформулируем свой запрос на период обучения. Это может быть смена работы, применение новых знаний на своём текущем рабочем месте, открытие бизнеса, публикация статьи или целой книги, что угодно! Цель может состоять из одного глобального результата или быть разбита на десяток мелких. Главное, чтобы её всегда можно было проверить по пяти критериям:
 - S (конкретность)	Что будет (произойдёт) в результате достижения этой цели?
 - M (измеримость)	Чем будем измеряться результат? В каких абсолютных величинах он будет выражаться?
 - A (достижимость)	Все ли необходимые ресурсы есть для достижения цели?
 - R (значимость)	Зачем это нужно? Почему это значимо?
 - T (ориентированность во времени)	Точные сроки старта и окончания твоих целей, а также промежуточные контрольные точки.
