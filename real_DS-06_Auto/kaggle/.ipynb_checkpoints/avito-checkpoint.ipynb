```python
# Импортируем наши библиотеки    
from bs4 import BeautifulSoup    
import requests
import urllib.request
from os import system
import json
import ssl
import codecs
import pandas as pd
import numpy as np
import re as re
import time
```

```python
def parse_tth(tth, descr, rec_id, price):
    global df_all
    d = {}
    
    try:
        try:
            for tag in tth.find_all('li'):
                prm = str(tag.text).split(':')
                d[str(prm[0]).strip()] = [str(prm[1]).strip()]
        except:
            return -1
        
        #Добавление/трансформация колонок
        d['price'] = price
        d['url'] = [tth_url]
        
        try:
            numberOfDoors = float(d['Количество дверей'][0])
        except:
            numberOfDoors = 4.0
        d['numberOfDoors'] = numberOfDoors
        
        try:
            d['bodyType'] = d['Тип кузова']
            if d['Тип кузова'][0] in ['внедорожник', 'хетчбэк', 'универсал', 'седан']:
                d['bodyType'] = [d['bodyType'][0]+' '+str(numberOfDoors)+' дв.']
        except:
            d['bodyType'] = None
    
        try:
            modelDate = float(re.search('\d{4}', d['Поколение'][0])[0])
        except:
            modelDate = None
        d['modelDate'] = modelDate
        
        try:
            productionDate = float(d['Год выпуска'])
        except:
            productionDate = modelDate
        d['productionDate'] = productionDate

        try:
            volume = re.search('\d\.\d', d['Модификация'][0])[0]
            d['vehicleConfiguration'] = d['Тип кузова'][0]+' '+d['Коробка передач'][0]+' '+volume
        except:
            volume = 0
            d['vehicleConfiguration'] = None

        try:
            if d['Коробка передач'][0] == 'автомат':
                d['vehicleTransmission'] = 'автоматическая'
            elif d['Коробка передач'][0] == 'механика':
                d['vehicleTransmission'] = 'механическая'
            elif d['Коробка передач'][0] == 'робот':
                d['vehicleTransmission'] = 'роботизированная'
            else:
                d['vehicleTransmission'] = 'other'
        except:
            d['vehicleTransmission'] = 'Other'

        if float(volume) > 0:
            d['engineDisplacement'] = str(volume)+' LTR'
        else:
            d['engineDisplacement'] = None

        try:
            power = re.search('\d{2,} ', d['Модификация'][0])[0]
        except:
            power = 0
        if float(volume) > 0:
            d['enginePower'] = str(power)+'N12'
        else:
            d['enginePower'] = None

        d['description'] = descr

        try:
            mileage = float(re.search('\d{0,}', d['Пробег'][0])[0])
        except:
            mileage = 0.0
        d['mileage'] = mileage
        
        try:
            d['Состояние'] = d['Состояние']
        except:
            d['Состояние'] = None
        
        try:
            driver_qnt = re.search('\d{1,}', d['Владельцев по ПТС'][0])[0]
        except:
            driver_qnt = '0'
        if int(driver_qnt) == 0:
            d['Владельцы'] = driver_qnt+' владельцев'
        elif int(driver_qnt) == 1:
            d['Владельцы'] = driver_qnt+' владелец'
        elif int(driver_qnt) == 2:
            d['Владельцы'] = driver_qnt+' владельца'
        elif int(driver_qnt) >= 3:
            d['Владельцы'] = driver_qnt+' или более'
    
        #Пока неоткуда брать данные
        d['ПТС'] = 'Оригинал'       #пример: Оригинал
        d['Таможня'] = 'Растаможен' #пример: Растаможен
        
        #Рассчитаем сред. срок относительно июня 2020г. и считая кол-во владельцев по ПТС
        if driver_qnt == 0:
            d['Владение'] = np.NaN #пример: NaN или "5 лет и 3 месяца"
        else:
            m_add = 6 #кол-во месяцев текущего года
            y_qnt = round(12*(2020.0 - productionDate) // 12 / driver_qnt)
            m_qnt = round((12*(2020.0 - productionDate) + m_add) % 12 / driver_qnt)
           
            y_qnt_str = 'лет'
            m_qnt_str = 'месяцев'
            
            if y_qnt == 1: y_qnt_str = 'год'
            elif 2 <= y_qnt <= 4: y_qnt_str = 'года'
            
            if m_qnt == 1: m_qnt_str = 'месяц'
            elif 2 <= m_qnt <= 4: m_qnt_str = 'месяца'
                
            if y_qnt == 0:
                d['Владение'] = str(m_qnt)+' '+m_qnt_str
            else:
                d['Владение'] = str(y_qnt)+' '+y_qnt_str+' '+str(m_qnt)+' '+m_qnt_str

        #Заполняем сквозной нумерацией
        d['id'] = rec_id

        #Приведение к формату файла test.csv
        df = pd.DataFrame(d)
        df_one = pd.concat(
            [
            df['bodyType'], df['Марка'], df['Цвет'], df['Тип двигателя'], df['modelDate'], 
            df['Модификация'], df['numberOfDoors'], df['productionDate'], df['vehicleConfiguration'],
            df['vehicleTransmission'], df['engineDisplacement'], df['enginePower'], 
            df['description'], df['mileage'], df['Комплектация'], df['Привод'], df['Руль'], 
            df['Состояние'], df['Владельцы'], df['ПТС'], df['Таможня'], df['Владение'],
            df['id'], df['price'], df['url']
            ],
            axis=1
        )
        df_one.columns=[
            'bodyType', 'brand', 'color', 'fuelType', 'modelDate', 
            'name', 'numberOfDoors', 'productionDate', 'vehicleConfiguration', 
            'vehicleTransmission', 'engineDisplacement', 'enginePower', 
            'description', 'mileage', 'Комплектация', 'Привод', 'Руль', 
            'Состояние', 'Владельцы', 'ПТС', 'Таможня', 'Владение',
            'id', 'price', 'url'
        ]

        df_all = df_all.append(df_one, ignore_index=True)
    except (KeyError, TypeError) as err:
        df_all.to_csv('avito_tmp_'+time.strftime('%Y%m%d%H%M%S', time.localtime())+'.csv')
        print(tth_url, ':', type(err), ': {}'.format(err.args[-1]))
    
    return 0
```

```python
def process_tth_url(tth_url, only_descr=False):
    if tth_url.find('automama') > 0: #ссылки на другой сайт игнорируем
        if only_descr:
            return -1
        else:
            return [-1,-1,-1]

    try: #Переходим на страницу с ТТХ
        tth_response = requests.get(tth_url, headers={'User-Agent': 'Mozilla/5.0'})    
        tth_page = BeautifulSoup(tth_response.text, 'html.parser')    
    except:
        if only_descr:
            return -1
        else:
            return [-1,-1,-1]

    try: #Описание про автомобиль лежит отдельно
        descr = tth_page.find('div', class_='item-description-text').text.strip()
    except:
        try:
            descr = tth_page.find('div', class_='item-description-html').text.strip()
        except:
            descr = None

    if only_descr:
        return descr
    else:
        tth = tth_page.find('ul', class_='item-params-list')

        try:
            #Цена также в отдельном месте
            price = tth_page.find('span', class_='js-item-price')
            price = float(price.attrs['content'])
        except AttributeError:
            price = 0

        return [tth, descr, price]

# process_tth_url('https://www.avito.ru/moskva/avtomobili/volkswagen_tiguan_2012_1867991470', 
#                 only_descr=True)
```

```python
url_city = {
    'Москва':'https://www.avito.ru/moskva/avtomobili?s=104&radius=0&p=',
    #'МосО':'https://www.avito.ru/moskovskaya_oblast/avtomobili?s=104&p=',
    #'Питер':'https://www.avito.ru/sankt-peterburg/avtomobili?s=104&radius=0&p=',
    #'ЛенО':'https://www.avito.ru/leningradskaya_oblast/avtomobili?s=104&p=',
    #'СверО':'https://www.avito.ru/sverdlovskaya_oblast/avtomobili?s=104&p=',
    #'РостО':'https://www.avito.ru/rostovskaya_oblast/avtomobili?s=104&p=',
    #'ЧелО':'https://www.avito.ru/chelyabinskaya_oblast/avtomobili?s=104&p=',
    #'КраснодК':'https://www.avito.ru/krasnodarskiy_kray/avtomobili?s=104&p=',
    #'СамО':'https://www.avito.ru/samarskaya_oblast/avtomobili?s=104&p='
} #осн. адрес, к нему будем добавлять номер страницы
```

```python
cnt_ads = 39623  #счетчик объявлений
cnt_err = 0  #счетчик ошибок
df_all = pd.DataFrame()

qnt4compare = 1000
is_debug = False

print(time.strftime('%d.%m.%Y %H:%M:%S', time.localtime())+\
      ': Парсинг автомобильных объявлений с сайта avito.ru ...')
if is_debug:
    print('!Отладка включена!')

for key in url_city.keys():
    if is_debug:
        if cnt_ads > qnt4compare:
            print('Прерываем цикл. cnt_ads =',cnt_ads)
            df_all.to_csv('avito_auto_'+str(cnt_ads)+'.csv')
            break
    print('     ===== Регион ['+key+'] =====')
    
    for cnt_page in range(1,101): #на сайте отображены только 100 страниц
        if is_debug:
            if cnt_ads > qnt4compare:
                break

        url = url_city[key]+str(cnt_page)
        try:
            response = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'})    
            page = BeautifulSoup(response.text, 'html.parser')
        except:
            print('Ошибка выполнения!')
            break

        #Получаем список объявлений на странице
        ads = page.find('div', class_='snippet-list js-catalog_serp')
        
        try:
            for ad in ads.find_all('a', class_='snippet-link'):
                if is_debug:
                    if cnt_ads > qnt4compare:
                        break

                tth_url = 'https://www.avito.ru'+ad.attrs['href']
                lst_prm = process_tth_url(tth_url)
                if lst_prm[0] != -1:
                    tth, descr, price = lst_prm[0], lst_prm[1], lst_prm[2]
                    #Парсинг данных
                    if parse_tth(tth=tth, descr=descr, rec_id=cnt_ads+1, price=price) == 0:
                        cnt_ads += 1
                        if cnt_ads % (qnt4compare/10) == 0:
                            print(time.strftime('%d.%m.%Y %H:%M:%S', time.localtime())+': Обработано', 
                                  cnt_ads, 'объявлений')
                        if cnt_ads % qnt4compare == 0:
                            df_all.to_csv('avito_auto_'+str(cnt_ads)+'_'+key+'.csv')
                            df_all = pd.DataFrame()

                time.sleep(1) #на всякий случай задержка
        except:
            continue
    #end for cnt_page
    
# Дописываем последние записи
if cnt_ads % qnt4compare != 0:
    df_all.to_csv('avito_auto_'+str(cnt_ads)+'.csv')
print('Всего обработано', cnt_ads, 'объявлений')
```
