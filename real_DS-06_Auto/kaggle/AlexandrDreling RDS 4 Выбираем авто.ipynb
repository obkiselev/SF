```python
import pandas as pd
from bs4 import BeautifulSoup   
import requests 
```

```python
# Загрузка главной страницы сайта
url_main = 'https://auto.drom.ru/used/all/'
response_1 = requests.get(url_main) 
page_1 = BeautifulSoup(response_1.text, 'html.parser')
```

```python
print(response_1)
```

```python
# Цикл сбора ссылок на автомобили со страницы 
c = page_1.find_all('a' , class_="b-advItem")
cars_html = []
for i in range (len(c)):
    cars_html.append(c[i].attrs['href'].strip())
```

```python
df = pd.DataFrame(columns = ['Name' , 'Двигатель', 'Мощность', 'Трансмиссия', 'Привод', 'Тип кузова', 'Цвет', 'Пробег, км', 'Руль' , 'Поколение' , 'Комплектация'])
```

```python
df
```

```python
n = 0
for link in cars_html:
    response = requests.get(link) 
    page = BeautifulSoup(response.text, 'html.parser')
    car = page.find_all('div' , class_ = 'b-media-cont_margin_b-size-xxs')
    d = {}
    d['Name'] = page.find('h1' , class_ = 'b-title b-title_type_h1 b-title_no-margin b-text-inline').text.strip()
    for i in range (len(car)):
        if i == 1 and car[i].find('span' , class_ = 'b-text-gray').text.replace(':' , '') == 'Мощность':
            d[car[i].find('span' , class_ = 'b-text-gray').text.replace(':' , '')] = car[i].find('div' , class_='b-link b-link_dashed').text.strip()
        else:
            d[car[i].find('span' , class_ = 'b-text-gray').text.replace(':' , '')] = car[i].find('span' , class_ = 'b-text-gray').next_element.next_element.strip()
    for item in d.items():
        df.loc[n , item[0]] = item[1]
    n += 1
```
