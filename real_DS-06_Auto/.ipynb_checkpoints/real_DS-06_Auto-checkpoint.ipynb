### Метрика

MAPE (mean percentage absolute error, http://en.wikipedia.org/wiki/Mean_absolute_percentage_error) — расшифровывается выражение как средняя абсолютная ошибка в процентах.
$$MAPE=\frac{1}{n}\sum_{t=1}^{n}\frac{\left | Y_t-\hat{Y_t} \right |}{Y_t}, где:$$
- $Y_t$ — фактическое значение за анализируемый период;
- $\hat{Y_t}$ — значение прогнозной модели за анализируемый период;
- $n$ — количество периодов.

Это коэффициент, не имеющий размерности, с очень простой интерпретацией. Его можно измерять в долях или процентах. Например, если у вас получилось, что MAPE=11.4%, то это говорит о том, что ошибка составила 11.4% от фактических значений. Вам нужно добиться минимального процента ошибки.

Чтоб понять, к какому уровню вам стоит стремиться — достаточно взглянуть на ЛидерБорд соревнования.
